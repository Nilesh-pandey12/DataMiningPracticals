Importing the Libraries

```python
import pandas as pd
```

Loading the wine dataset

```python
dataset=pd.read_csv('winequalityN.csv')
```

```python
dataset.head(5)
```

```python
dataset.shape
```

Statistical Description of Data

```python
desc=dataset.describe()
desc
```

Mean and Standard Deviation of Data

```python
wineStats=desc.loc[['mean','std']]
wineStats
```

Checking if columns are standardized

```python
for i in wineStats.columns:
    if(wineStats[i]['mean']!=0 and wineStats[i]['std']!=1):
        print(i)
```

Dropping the string type column

```python
dataset.drop(columns='type',inplace=True)
```

```python
dataset
```

Standardizing the data

```python
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

standardData = scaler.fit_transform(dataset) 
print(standardData)
```

Statistical Description of Standardized data

```python
pd.DataFrame(standardData,columns=dataset.columns).describe()
```

```python
import seaborn as sns
```

Loading the iris dataset

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

Standardizing the data

```python
from sklearn.preprocessing import StandardScaler

scaler_iris= StandardScaler()

data=iris.iloc[:,:-1]

standardirisData = scaler_iris.fit_transform(data) 
print(standardirisData)
```

Statistical Description of Standardized data

```python
pd.DataFrame(standardirisData,columns=data.columns).describe()
```
